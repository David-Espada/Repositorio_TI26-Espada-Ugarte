# 1. Exploración Inicial del Dataset


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Revenue")
plt.title("Distribución de compras realizadas")
plt.show()
```

```python
# porcentaje de compras
df["Revenue"].value_counts(normalize=True) * 100
```

```python
#tipo de visitante
plt.figure(figsize=(8,4))
sns.countplot(data=df, x="VisitorType")
plt.title("Distribución por tipo de visitante")
plt.xticks(rotation=15)
plt.show()
```

```python
#mes de la visita
plt.figure(figsize=(10,4))
sns.countplot(data=df, x="Month")
plt.title("Sesiones por mes")
plt.show()
```

```python
#matriz de correlacion
plt.figure(figsize=(12,8))

corr = df.select_dtypes(include=["int64","float64"]).corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Matriz de correlación")
plt.show()
```

```python
df["Revenue"].value_counts(normalize=True) * 100
```

```python
#histograma PageVaues
plt.figure(figsize=(8,5))
sns.histplot(df["PageValues"], bins=30)
plt.title("Distribución de PageValues")
plt.show()
```
